```python
fig, ax1 = plt.subplots(figsize=(12, 9))
ax1.spines["top"].set_visible(False)  
ax1.spines["left"].set_visible(False) 
ax1.spines["right"].set_visible(False) 
ax1.plot(tempYears, sunshine, 'b--')
ax1.plot(tempYears, sunshine, 'bo')
ax1.set_xlabel('Time (years)', fontsize=10)
# Make the y-axis label, ticks and tick labels match the line color.
ax1.set_ylabel('sunny hours', color='b', fontsize=16)
ax1.tick_params('y', colors='b')
ax1.spines["bottom"].set_visible(False)  
ax2 = ax1.twinx()
ax2.plot(tempYears, rainyDays, '--', color='#ad9c08')
ax2.plot(tempYears, rainyDays, 'o', color='#ad9c08')
ax2.set_ylabel('rainy days', color='#ad9c08', fontsize=16)
ax2.tick_params('y', colors='#ad9c08')

ax2.spines["top"].set_visible(False) 
ax2.spines["bottom"].set_visible(False) 
ax2.spines["left"].set_visible(False)
ax2.spines["right"].set_visible(False)
# Make the title big enough so it spans the entire plot, but don't make it  
# so big that it requires two lines to show.  
plt.title("Temperature change throughout years", fontsize=22) 

plt.show()
```
