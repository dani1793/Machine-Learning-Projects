```python
from __future__ import print_function
from keras.applications.inception_v3 import InceptionV3
from keras.preprocessing import image
from keras.models import Model
from keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.python.keras.preprocessing.image import ImageDataGenerator
from keras import backend as K
from keras.datasets import cifar10
from keras.layers import Input
import keras
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load the data
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

Convert data into numpy arrrays so that it could be used as an input vector

```python
# convert the data into desired format
x_train = keras.backend.spatial_2d_padding(x_train, padding=((60, 60), (60, 60)), data_format=None)
x_test = keras.backend.spatial_2d_padding(x_test, padding=((60, 60), (60, 60)), data_format=None)
x_train = K.cast(x_train, dtype='float32')
x_test = K.cast(x_test, dtype='float32')

sess = tf.Session()
with sess.as_default():
    x_train = x_train.eval()
    x_test = x_test.eval()
sess.close()
```

Initialize the basic Inception model trained on imagenet weights

```python
# this could also be the output a different Keras model or layer
# create the base pre-trained model
base_model = InceptionV3(weights='imagenet', include_top=False)
```

Add a layer on top of the Inception model to classify the image set

```python
# add a global spatial average pooling layer
x = base_model.output
x = GlobalAveragePooling2D()(x)
# let's add a fully-connected layer
x = Dense(1024, activation='relu')(x)
# and a logistic layer -- let's say we have 200 classes
predictions = Dense(10, activation='softmax')(x)

# convert class vectors to binary class matrices
y_train_cat = keras.utils.to_categorical(y_train, 10)
y_test_cat = keras.utils.to_categorical(y_test, 10)
```

Fit the model without any preprocessing. The model is trained for only two epochs due to limited computation power

```python

# this is the model we will train
model = Model(inputs=base_model.input, outputs=predictions)

# first: train only the top layers (which were randomly initialized)
# i.e. freeze all convolutional InceptionV3 layers
for layer in base_model.layers:
    layer.trainable = False

metrics = ['categorical_accuracy']

# compile the model (should be done *after* setting layers to non-trainable)
model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=metrics)
# train the model on the new data for a few epochs
hist = model.fit(x_train, 
                    y_train_cat, 
                    epochs=2, 
                    batch_size=20, 
                    validation_data = (x_test, y_test_cat),
                    verbose=0)
```

save the weights for future use

```python
model.save_weights('./inception_weights.hdf5')
```

Introduce normalization to the model

```python

# this is the model we will train
norm_model = Model(inputs=base_model.input, outputs=predictions)

# first: train only the top layers (which were randomly initialized)
# i.e. freeze all convolutional InceptionV3 layers
for layer in base_model.layers:
    layer.trainable = False

metrics = ['categorical_accuracy']

# compile the model (should be done *after* setting layers to non-trainable)
norm_model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=metrics)


datagen_train = ImageDataGenerator(
      rescale=1./255,
      fill_mode='nearest')

datagen_test = ImageDataGenerator(rescale=1./255)

norm_hist = norm_model.fit_generator(datagen_train.flow(x_train, y_train_cat, batch_size=20),
                                     steps_per_epoch=len(x_train) / 20, 
                                     epochs=2, 
                                     validation_data = datagen_test.flow(x_test, y_test_cat, batch_size=20),
                                     validation_steps = len(x_test) / 20,
                                     verbose=0)
```

Save the model weights with normalized input

```python
norm_model.save_weights('./inception_weights-norm.hdf5')
```

Compare the results of both un-normalized and normalized model

```python
plt.clf()
plt.plot(norm_hist.history['val_loss'], label='Val Loss Normalized')
plt.scatter([0, 1], norm_hist.history['val_loss'])

plt.plot(hist.history['val_loss'], label='Val Loss')
plt.scatter([0, 1], hist.history['val_loss'])
plt.title('Validation Loss')
plt.xlabel('epochs')
plt.legend()
plt.show()

plt.clf()
plt.plot(norm_hist.history['val_categorical_accuracy'], label='Val Cat Acc Norm')
plt.scatter([0, 1], norm_hist.history['val_categorical_accuracy'])
plt.plot(hist.history['val_categorical_accuracy'], label='Val Cat Acc')
plt.scatter([0, 1], hist.history['val_categorical_accuracy'])
plt.title('Validation Categorical Accuracy')
plt.xlabel('epochs')
plt.legend()
plt.show()

plt.clf()

plt.plot(norm_hist.history['loss'], label='Loss Normalized')
plt.scatter([0, 1], norm_hist.history['loss'])
plt.plot(hist.history['loss'], label='Loss')
plt.scatter([0, 1], hist.history['loss'])
plt.title('Loss')
plt.xlabel('epochs')
plt.legend()
plt.show()

plt.clf()
plt.plot(norm_hist.history['categorical_accuracy'], label='Cat Acc Norm')
plt.scatter([0, 1], norm_hist.history['categorical_accuracy'])
plt.plot(hist.history['categorical_accuracy'], label='Cat Acc')
plt.scatter([0, 1], hist.history['categorical_accuracy'])
plt.title('Categorical Accuracy')
plt.xlabel('epochs')
plt.legend()
plt.show()
```
